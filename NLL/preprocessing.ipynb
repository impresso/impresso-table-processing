```python
import os
import sys
import json
import numpy as np
import pandas as pd

sys.path.append("../helpers/")
from impresso_id import *

data_path = "data"

%load_ext autoreload
%autoreload 2
```

# Pre-processing

```python
with (open(os.path.join(data_path, "NLL_metadata.json"), "r")) as f:
    NLL_metadata = json.load(f)
```

## NLL

```python
NLL = pd.DataFrame().from_dict(NLL_metadata, orient="index")
print(NLL.shape)
NLL.head()
```

```python
NLL["year"] = NLL["pid"].apply(lambda x: get_year(x))
NLL["month"] = NLL["pid"].apply(lambda x: get_month(x))
NLL["day"] = NLL["pid"].apply(lambda x: get_day(x))
NLL["journal"] = NLL["pid"].apply(lambda x: get_journal(x))
NLL["edition"] = NLL["pid"].apply(lambda x: get_edition(x))
NLL["page"] = NLL["pid"].apply(lambda x: get_page(x))
NLL["meta_issue_id"] = NLL["pid"].apply(lambda x: get_meta_issue_id(x))

NLL["x"] = NLL["tb_coords"].apply(lambda x: np.min(x["x"]))
NLL["y"] = NLL["tb_coords"].apply(lambda x: np.min(x["y"]))
NLL["width"] = NLL["tb_coords"].apply(lambda x: np.max([x + w for x, w in zip(x['x'], x['width'])]) - np.min(x["x"]))
NLL["height"] = NLL["tb_coords"].apply(lambda x: np.max([x + h for x, h in zip(x['y'], x['height'])]) - np.min(x["y"]))
NLL["area"] = NLL.apply(lambda x: x["width"]*x["height"], axis=1)

NLL.head()
```

When fetching the images from the IIIF of the National Library of Luxembourg, it appeared that one link was dead and therefore the image couldn't be retrieved. Hence the need to remove it manually from the dataset.

```python
NLL.loc["luxwort-1936-04-22-a-i0079", "iiifURL"]
```

```python
NLL = NLL.drop(list(NLL.loc[NLL['pid'] == NLL.loc["luxwort-1936-04-22-a-i0079", 'pid']].index))
```

## NLL-tag

```python
with (open("NLL_tag.json", "r")) as f:
    NLL_tag = json.load(f)
```

```python
NLL_tag = pd.DataFrame().from_dict(NLL_tag, orient="index").join(NLL)
print(NLL_tag.shape)
NLL_tag.head()
```

```python
NLL_tag.loc[NLL_tag['tag'] == 'table', 'tag'] = 'miscellaneous'
NLL_tag.loc[NLL_tag['tag'] == 'sports results', 'tag'] = 'sport results'
```

## NLL-auto

```python
with (open("NLL_auto.json", "r")) as f:
    NLL_auto = json.load(f)
```

```python
NLL_auto = pd.DataFrame().from_dict(NLL_auto, orient="index").join(NLL)
print(NLL_auto.shape)
NLL_auto.head()
```

```python
NLL_auto.loc[NLL_auto['tag'] == 'table', 'tag'] = 'miscellaneous'
NLL_auto.loc[NLL_auto['tag'] == 'sports results', 'tag'] = 'sport results'
```
